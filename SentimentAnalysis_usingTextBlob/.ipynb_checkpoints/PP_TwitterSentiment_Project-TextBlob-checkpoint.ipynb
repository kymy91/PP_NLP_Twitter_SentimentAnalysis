# Sentiment analysis of Twitter tweets using Natural Language Processing
Sentiment analysis using Natural Language Processing (TextBlob) and WordCloud for data visualization.

Analysing Subjectivity and Polarity. By identifying which  tweets as positive, negative and neutral.
Use the WordCloud to display a word cloud of the most common words analysed.

```python
import pandas as pd
import numpy as np
import re

# TextBlob - Python library for processing textual data
from textblob import TextBlob
# WordCloud - Python linrary for creating image wordclouds
from wordcloud import WordCloud

# Data Viz
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 12, 8
```

-----
Pre-Processing

```python
def prepro_data(df):
    df=df.set_index('id')
    df=df.drop(['Unnamed: 0'], axis = 1)
    return df
```

```python
def prepro_tweets(txt): 

        
    #remove_hyperlink
    txt['tweetcleaned'] = txt['tweet'].apply(lambda x: re.sub(r"http\S+",'', str(x)))
    #remove_mentions
    txt['tweetcleaned'] = txt['tweet'].apply(lambda x: re.sub(r"@\S+",'', str(x)))
    #remove_number
    txt['tweetcleaned'] = txt['tweet'].apply(lambda x: re.sub(r'\d+','', str(x)))
    #remove_punctuation
    txt['tweetcleaned'] = txt['tweet'].apply(lambda x: re.sub('[^A-Za-z]+',' ', str(x)))
    
    #to lower
    txt['tweetcleaned'] = list(map(lambda x: x.lower(), txt['tweetcleaned']))
    #remove_whitespace
    txt['tweetcleaned'] = list(map(lambda x: x.strip(), txt['tweetcleaned']))
    #replace_newline(word)
    txt['tweetcleaned'] = list(map(lambda x: x.replace('\n', ''), txt['tweetcleaned']))
    #remove user
    txt['tweetcleaned'] = list(map(lambda x: x.replace('user', ''), txt['tweetcleaned']))
                                       
    return txt
    
```

```python
data=prepro_data(data)
```

```python
prepro_tweets(data)              
```

```python
def getTextSubjectivity(txt):
    return TextBlob(txt).sentiment.subjectivity

def getTextPolarity(txt):
    return TextBlob(txt).sentiment.polarity
```

```python
# negative, nautral, positive analysis
def TextAnalysis(a):
    if a < 0:
        return "Negative"
    elif a == 0:
        return "Neutral"
    else:
        return "Positive"
```

```python
data['Subjectivity'] = data['tweetcleaned'].apply(getTextSubjectivity)
data['Polarity'] = data['tweetcleaned'].apply(getTextPolarity)
data
```

```python
data['TextAnalysis_Score'] = data['Polarity'].apply(TextAnalysis)
data
```

Analysis

```python
positive = data[data['TextAnalysis_Score'] == 'Positive']

print(str(positive.shape[0]/(data.shape[0])*100) + " % of positive tweets")
```

```python
objective = data[data['Subjectivity'] == 0]

print(str(objective.shape[0]/(data.shape[0])*100) + " % of objective tweets")
```

```python
labels = data.groupby('TextAnalysis_Score').count().index.values

values = data.groupby('TextAnalysis_Score').size().values
plt.title('Twitter Sentiment Analysis Score Bar chart')
plt.bar(labels, values)
```

```python
for index, row in data.iterrows():
    if row['TextAnalysis_Score'] == 'Positive':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="green")
    elif row['TextAnalysis_Score'] == 'Negative':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="red")
    elif row['TextAnalysis_Score'] == 'Neutral':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="blue")

plt.title('Twitter Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```

```python
# word cloud of the most common words
words = ' '.join([tweet for tweet in data['tweetcleaned']])
wordCloud = WordCloud(width=600, height=400).generate(words)

plt.imshow(wordCloud)
plt.show()
```
